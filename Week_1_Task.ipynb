```python
def interquartile(data):
    result = {}
    for col in data.columns:
        print("Looking at " +col)
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5*IQR
        upper = Q3 + 1.5*IQR
        print(lower)
        print(upper)

        # Create arrays of Boolean values indicating the outlier rows
        upper_array = np.where(data[col] >= upper)[0]
        lower_array = np.where(data[col] <= lower)[0]
        total_array = np.concatenate((upper_array,lower_array))
        print(upper_array.shape)
        print(lower_array.shape)

        result[col] = np.shape(total_array)
        print(total_array.shape)
        print("done with " + col)
        break
    return result
```

```python
outlier_array = interquartile(df)
```
